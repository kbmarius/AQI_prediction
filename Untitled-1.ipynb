III. DATA SPLITING ET CREATION DES VARIABLES

III.1 Création de la variable cible AQI et sa classification

```python
# Création de la cible AQI à partir de CO et NO2
def calculate_aqi(co, no2):
    co_aqi = co / 10 * 50       # Normalisation CO: 0-10 mg/m³ → 0-50 AQI
    no2_aqi = no2 / 100 * 50    # Normalisation NO2: 0-100 µg/m³ → 0-50 AQI
    return np.maximum(co_aqi, no2_aqi)

df["AQI"] = calculate_aqi(df["CO(GT)"], df["NO2(GT)"])
```

```python
# Classification de l'AQI
def classify_aqi(aqi_value):
    """Classification de la qualité de l'air basée sur l'AQI"""
    if aqi_value <= 25:
        return "Excellent"
    elif aqi_value <= 50:
        return "Bon"
    elif aqi_value <= 75:
        return "Moyen"
    elif aqi_value <= 100:
        return "Médiocre"
    else:
        return "Dangereux"

df['AQI_Class'] = df['AQI'].apply(classify_aqi)
```

III.2 Feature engeneering

```python

# Features temporelles
df['hour'] = df['Datetime'].dt.hour
df['day_of_week'] = df['Datetime'].dt.dayofweek
df['month'] = df['Datetime'].dt.month
df['season'] = df['month'] % 12 // 3 + 1
df['weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

# Features supplémentaires
df['CO_NO2_ratio'] = df['CO(GT)'] / (df['NO2(GT)'] + 1)  # +1 pour éviter la division par zéro
df['temp_humidity'] = df['T'] * df['RH']
```

III.3 Créations des variables

```python
#Features: capteurs et paramètres environnementaux et features crées
features = [
    'PT08.S1(CO)', 'PT08.S2(NMHC)', 'PT08.S3(NOx)', 'PT08.S4(NO2)', 'PT08.S5(O3)',
    'T', 'RH', 'AH', 'hour', 'day_of_week', 'month', 'season', 'weekend',
    'CO_NO2_ratio', 'temp_humidity'
]

X = df[features] #variables explicatives
y = df["AQI"]  #variable expliquée
```

III.4 Normalisation et division train/test

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
   X_scaled, y, test_size=0.2, random_state=42
)
```

```python
print("Taille du jeu d'entraînement :", X_train.shape[0])
print("Taille du jeu de test        :", X_test.shape[0])
```

IV. MODELISATION

IV.1 Entrainement des modèles

```python
results = {}

# -------- Régression linéaire --------
print("Entraînement de Linear Regression...")
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

mse = mean_squared_error(y_test, y_pred_lr)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred_lr)
r2 = r2_score(y_test, y_pred_lr)
mape = np.mean(np.abs((y_test - y_pred_lr) / y_test)) * 100

results['Linear Regression'] = {
    'model': lr,
    'mse': mse,
    'rmse': rmse,
    'mae': mae,
    'r2': r2,
    'mape': mape
}

print(f"Linear Regression - R²: {r2:.3f}, RMSE: {rmse:.3f}, MAE: {mae:.3f}, MAPE: {mape:.2f}%")

# -------- Random Forest --------
print("\nEntraînement de Random Forest...")
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

mse = mean_squared_error(y_test, y_pred_rf)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)
mape = np.mean(np.abs((y_test - y_pred_rf) / y_test)) * 100

results['Random Forest'] = {
    'model': rf,
    'mse': mse,
    'rmse': rmse,
    'mae': mae,
    'r2': r2,
    'mape': mape
}

print(f"Random Forest - R²: {r2:.3f}, RMSE: {rmse:.3f}, MAE: {mae:.3f}, MAPE: {mape:.2f}%")

# -------- Gradient Boosting --------
print("\nEntraînement de Gradient Boosting...")
gb = GradientBoostingRegressor(random_state=42)
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)

mse = mean_squared_error(y_test, y_pred_gb)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred_gb)
r2 = r2_score(y_test, y_pred_gb)
mape = np.mean(np.abs((y_test - y_pred_gb) / y_test)) * 100

results['Gradient Boosting'] = {
    'model': gb,
    'mse': mse,
    'rmse': rmse,
    'mae': mae,
    'r2': r2,
    'mape': mape
}

print(f"Gradient Boosting - R²: {r2:.3f}, RMSE: {rmse:.3f}, MAE: {mae:.3f}, MAPE: {mape:.2f}%")

# -------- SVR --------
print("\nEntraînement de SVR...")
svr = SVR(kernel='rbf')
svr.fit(X_train, y_train)
y_pred_svr = svr.predict(X_test)

mse = mean_squared_error(y_test, y_pred_svr)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred_svr)
r2 = r2_score(y_test, y_pred_svr)
mape = np.mean(np.abs((y_test - y_pred_svr) / y_test)) * 100

results['SVR'] = {
    'model': svr,
    'mse': mse,
    'rmse': rmse,
    'mae': mae,
    'r2': r2,
    'mape': mape
}

print(f"SVR - R²: {r2:.3f}, RMSE: {rmse:.3f}, MAE: {mae:.3f}, MAPE: {mape:.2f}%")
```

```python
#Regréssion linéaire
# --- 1) Scatter plot : Observé vs Prédit ---
def plot_obs_vs_pred(y_true, y_pred, title="Régression linéaire – Observé vs Prédit"):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    fig, ax = plt.subplots(figsize=(7, 5))
    ax.scatter(y_true, y_pred, s=20, color="blue", alpha=0.6, label="Points")
    
    # Diagonale de référence (y=x)
    vmin = np.nanmin([y_true.min(), y_pred.min()])
    vmax = np.nanmax([y_true.max(), y_pred.max()])
    ax.plot([vmin, vmax], [vmin, vmax], linestyle="--", color="red", linewidth=1.5, label="y = x")

    ax.set_title(title)
    ax.set_xlabel("AQI observé")
    ax.set_ylabel("AQI prédit")
    ax.legend()
    plt.tight_layout()
    plt.show()

# --- 2) Graphique des résidus ---
def plot_residuals_vs_pred(y_true, y_pred, title="Régression linéaire – Résidus vs Prédit"):
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    resid  = y_true - y_pred

    fig, ax = plt.subplots(figsize=(7, 5))
    ax.scatter(y_pred, resid, s=20, color="blue", alpha=0.6, label="Résidus")
    
    # Ligne horizontale y=0
    ax.axhline(0, linestyle="--", color="red", linewidth=1.5)

    ax.set_title(title)
    ax.set_xlabel("AQI prédit")
    ax.set_ylabel("Résidu (observé - prédit)")
    plt.tight_layout()
    plt.show()


plot_obs_vs_pred(y_test, y_pred_lr)
plot_residuals_vs_pred(y_test, y_pred_lr)
```

```python

def safe_mape(y_true, y_pred):
    """
    Mean Absolute Percentage Error (MAPE) with protection
    against division by zero.
    """
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mask = y_true != 0  
    if not np.any(mask):
        return np.nan  
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100
```

IV.3 Grrid Search (Optimisation des hyperparamètres du meilleur modèle "Random Forest")

```python
# (1) utilitaire MAPE robuste (au cas où on veut la reporter)
def safe_mape(y_true, y_pred):
    y_true = np.asarray(y_true); y_pred = np.asarray(y_pred)
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask]-y_pred[mask]) / y_true[mask]))*100 if np.any(mask) else np.nan

# (2) results existe ? sinon on l'initialise
try:
    results
except NameError:
    results = {}

# (3) Estimateur et validation croisée temporelle
rf = RandomForestRegressor(random_state=42, n_jobs=-1)
cv = TimeSeriesSplit(n_splits=5)   # important pour données chronologiques

# (4) Grille d’hyperparamètres
param_grid = {
    "n_estimators": [200, 400, 600],
    "max_depth": [None, 20, 30],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "max_features": ["sqrt", "log2", 1.0] 
}

# (5) GridSearchCV : on optimise la RMSE
gs = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=cv,
    scoring="neg_root_mean_squared_error",  # minimise la RMSE
    n_jobs=-1,
    verbose=2,
    refit=True
)

print(">> GridSearch Random Forest (TimeSeriesSplit, scoring=RMSE)...")
gs.fit(X_train, y_train)

print("\nMeilleurs hyperparamètres :", gs.best_params_)
print(f"Meilleur RMSE (CV) ≈ {abs(gs.best_score_):.4f}")

# (6) Meilleur modèle et évaluation sur TEST
best_model = gs.best_estimator_
y_pred_best = best_model.predict(X_test)

mse_best  = mean_squared_error(y_test, y_pred_best)
rmse_best = np.sqrt(mse_best)
mae_best  = mean_absolute_error(y_test, y_pred_best)
r2_best   = r2_score(y_test, y_pred_best)
mape_best = safe_mape(y_test, y_pred_best)

print("\nPerformances sur test (RF optimisé) :")
print(f"R²   = {r2_best:.3f}")
print(f"RMSE = {rmse_best:.3f}")
print(f"MAE  = {mae_best:.3f}")
print(f"MAPE = {mape_best:.2f}%")

# (7) Stockage SANS écraser le baseline
results["Random Forest (GridSearch)"] = {
    "model": best_model,
    "y_pred": y_pred_best,
    "mse": mse_best, "rmse": rmse_best, "mae": mae_best, "r2": r2_best, "mape": mape_best,
    "best_params": gs.best_params_
}

# (8) Comparaison baseline vs optimisé 
rows = []
if "Random Forest" in results:
    rows.append({
        "Modèle": "Random Forest (baseline)",
        "R²": results["Random Forest"]["r2"],
        "RMSE": results["Random Forest"]["rmse"],
        "MAE": results["Random Forest"]["mae"],
        "MAPE (%)": results["Random Forest"].get("mape", np.nan)
    })
rows.append({
    "Modèle": "Random Forest (GridSearch)",
    "R²": r2_best, "RMSE": rmse_best, "MAE": mae_best, "MAPE (%)": mape_best
})

df_rf_compare = pd.DataFrame(rows).sort_values(by="R²", ascending=False).reset_index(drop=True)
print("\nComparaison Random Forest (baseline vs optimisé) :")
print(df_rf_compare.round(3))

# (9) Sauvegarde du meilleur modèle (pour l'app Streamlit)
joblib.dump(best_model, "best_model.pkl")
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(features, 'features.pkl')
print("Modèle optimisé sauvegardé : best_model.pkl")
```

IV.6.1 Comparaison de l'AQI Réel vs AQI prédit moyen mensuel

```python
# === Heatmaps comparaison AQI mensuel (Réel vs Prédit) — valeurs + classes ===
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.patches import Patch

# Vérifs minimum
if 'df' not in globals() or 'classify_aqi' not in globals():
    raise NameError("Assure-toi d'avoir 'df' (avec 'Datetime' et 'AQI') et la fonction classify_aqi(...).")
if 'Datetime' not in df.columns or 'AQI' not in df.columns:
    raise ValueError("df doit contenir 'Datetime' et 'AQI'.")
if 'df_pred_mensuel' not in globals():
    raise NameError("df_pred_mensuel est introuvable. Crée-le d'abord (AQI_Predit_Moyen par mois).")

# 1) Réel mensuel (moyenne + classe)
dfr = df[['Datetime','AQI']].dropna().copy()
dfr['Datetime'] = pd.to_datetime(dfr['Datetime'])
real_m = (dfr.set_index('Datetime')
             .resample('M')['AQI'].mean()
             .to_frame('AQI_Reel_Moyen')
             .reset_index())
real_m['Année'] = real_m['Datetime'].dt.year
real_m['Mois_num'] = real_m['Datetime'].dt.month
real_m['Classe_Reelle'] = real_m['AQI_Reel_Moyen'].apply(classify_aqi)

# 2) Prédit mensuel (moyenne + classe)
dfp = df_pred_mensuel.copy()
if 'Datetime' in dfp.columns:
    dtp = pd.to_datetime(dfp['Datetime'])
elif 'Mois' in dfp.columns:
    dtp = pd.to_datetime(dfp['Mois'] + '-01', format='%Y-%m-%d', errors='coerce')
    dfp['Datetime'] = dtp
else:
    raise ValueError("df_pred_mensuel doit contenir 'Datetime' ou 'Mois' (AAAA-MM).")

if 'AQI_Predit_Moyen' not in dfp.columns:
    raise ValueError("df_pred_mensuel doit contenir 'AQI_Predit_Moyen'.")
if 'Interpretation' not in dfp.columns:
    dfp['Interpretation'] = dfp['AQI_Predit_Moyen'].apply(classify_aqi)

dfp['Année'] = pd.to_datetime(dfp['Datetime']).dt.year
dfp['Mois_num'] = pd.to_datetime(dfp['Datetime']).dt.month
pred_m = dfp[['Année','Mois_num','AQI_Predit_Moyen','Interpretation']].rename(columns={'Interpretation':'Classe_Predite'})

# 3) Grille commune Année × Mois (1..12)
years = sorted(set(real_m['Année']).union(set(pred_m['Année'])))
months = list(range(1, 12+1))

pivot_real_val = real_m.pivot(index='Année', columns='Mois_num', values='AQI_Reel_Moyen').reindex(index=years, columns=months)
pivot_real_cls = real_m.pivot(index='Année', columns='Mois_num', values='Classe_Reelle').reindex(index=years, columns=months)

pivot_pred_val = pred_m.pivot(index='Année', columns='Mois_num', values='AQI_Predit_Moyen').reindex(index=years, columns=months)
pivot_pred_cls = pred_m.pivot(index='Année', columns='Mois_num', values='Classe_Predite').reindex(index=years, columns=months)

# 4) Classes → couleurs fixes
palette_classes = {
    "Excellent": "#2ECC71",
    "Bon":       "#A3E4D7",
    "Moyen":     "#F1C40F",
    "Médiocre":  "#E67E22",
    "Dangereux": "#E74C3C",
}
class_order = list(palette_classes.keys())
class_to_num = {cls: i for i, cls in enumerate(class_order)}
cmap = mcolors.ListedColormap([palette_classes[c] for c in class_order])

num_real = pivot_real_cls.replace(class_to_num)
num_pred = pivot_pred_cls.replace(class_to_num)

# 5) Annotations "valeur + classe"
def make_annot(val_pivot, cls_pivot):
    arr = np.empty(val_pivot.shape, dtype=object)
    for i in range(val_pivot.shape[0]):
        for j in range(val_pivot.shape[1]):
            v = val_pivot.iat[i, j]
            c = cls_pivot.iat[i, j]
            arr[i, j] = "" if (pd.isna(v) or pd.isna(c)) else f"{v:.1f}\n{c}"
    return arr

annot_real = make_annot(pivot_real_val, pivot_real_cls)
annot_pred = make_annot(pivot_pred_val, pivot_pred_cls)

# 6) Tracé (deux heatmaps empilées, même échelle/légende)
mois_labels = ['Jan','Fév','Mar','Avr','Mai','Juin','Juil','Août','Sep','Oct','Nov','Déc']

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 12), sharex=True, sharey=True)

sns.heatmap(
    num_real, annot=annot_real, fmt="",
    cmap=cmap, cbar=False, linewidths=.3, linecolor='white',
    xticklabels=mois_labels, yticklabels=True, ax=axes[0]
)
axes[0].set_title("AQI mensuel RÉEL (moyenne + classe)")
axes[0].set_xlabel("")
axes[0].set_ylabel("Année")

sns.heatmap(
    num_pred, annot=annot_pred, fmt="",
    cmap=cmap, cbar=False, linewidths=.3, linecolor='white',
    xticklabels=mois_labels, yticklabels=True, ax=axes[1]
)
axes[1].set_title("AQI mensuel PRÉDIT (moyenne + classe)")
axes[1].set_xlabel("Mois")
axes[1].set_ylabel("Année")

# Légende commune
legend_handles = [Patch(facecolor=palette_classes[cls], label=cls) for cls in class_order]
axes[0].legend(handles=legend_handles, title="Classe AQI",
               bbox_to_anchor=(1.02, 1), loc='upper left', frameon=False)

plt.tight_layout()
plt.show()
```

```python
# === Heatmap des écarts AQI mensuels (Prédit − Réel) sur toute la période ===
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 0) Vérifs
if 'df' not in globals():
    raise NameError("Le DataFrame 'df' (avec 'Datetime' et 'AQI') est requis.")
if 'Datetime' not in df.columns or 'AQI' not in df.columns:
    raise ValueError("'df' doit contenir les colonnes 'Datetime' et 'AQI'.")
if 'df_pred_mensuel' not in globals():
    raise NameError("'df_pred_mensuel' (avec l'AQI mensuel prédit) est requis.")
if 'AQI_Predit_Moyen' not in df_pred_mensuel.columns:
    raise ValueError("'df_pred_mensuel' doit contenir la colonne 'AQI_Predit_Moyen'.")

# 1) Mensualisation du réel
dfr = df[['Datetime','AQI']].dropna().copy()
dfr['Datetime'] = pd.to_datetime(dfr['Datetime'])
real_m = (dfr.set_index('Datetime')
            .resample('M')['AQI'].mean()
            .to_frame('AQI_Reel_Moyen')
            .reset_index())
real_m['Année'] = real_m['Datetime'].dt.year
real_m['Mois_num'] = real_m['Datetime'].dt.month

# 2) Normalisation du prédictif
dfp = df_pred_mensuel.copy()
if 'Datetime' in dfp.columns:
    dfp['Datetime'] = pd.to_datetime(dfp['Datetime'])
elif 'Mois' in dfp.columns:
    dfp['Datetime'] = pd.to_datetime(dfp['Mois'] + '-01', format='%Y-%m-%d', errors='coerce')
else:
    raise ValueError("'df_pred_mensuel' doit contenir 'Datetime' ou 'Mois' (AAAA-MM).")

dfp['Année'] = dfp['Datetime'].dt.year
dfp['Mois_num'] = dfp['Datetime'].dt.month
pred_m = dfp[['Année','Mois_num','AQI_Predit_Moyen']].dropna(subset=['Année','Mois_num'])

# 3) Jointure sur la grille temporelle commune
years = sorted(set(real_m['Année']).union(set(pred_m['Année'])))
months = list(range(1, 13))

grid = (pd.MultiIndex.from_product([years, months], names=['Année','Mois_num'])
        .to_frame(index=False))

merged = (grid.merge(real_m[['Année','Mois_num','AQI_Reel_Moyen']], on=['Année','Mois_num'], how='left')
               .merge(pred_m, on=['Année','Mois_num'], how='left'))

# 4) Écart et erreur absolue
merged['Ecart'] = merged['AQI_Predit_Moyen'] - merged['AQI_Reel_Moyen']
merged['AbsErr'] = merged['Ecart'].abs()

# 5) Pivots pour heatmaps
pivot_diff = merged.pivot(index='Année', columns='Mois_num', values='Ecart')
pivot_abs  = merged.pivot(index='Année', columns='Mois_num', values='AbsErr')

# 6) Bornes symétriques autour de 0 (robustes)
max_abs = np.nanpercentile(np.abs(pivot_diff.values), 95)  # robuste aux outliers
if not np.isfinite(max_abs) or max_abs == 0:
    max_abs = np.nanmax(np.abs(pivot_diff.values)) if np.isfinite(np.nanmax(np.abs(pivot_diff.values))) else 1.0
vlim = float(max_abs)

# 7) Labels des mois
mois_labels = ['Jan','Fév','Mar','Avr','Mai','Juin','Juil','Août','Sep','Oct','Nov','Déc']
pivot_diff.columns = [mois_labels[m-1] for m in pivot_diff.columns]
pivot_abs.columns  = [mois_labels[m-1] for m in pivot_abs.columns]

# 8) Annotations ±x.x (pour l'écart)
annot = pivot_diff.copy()
for i in range(annot.shape[0]):
    for j in range(annot.shape[1]):
        val = annot.iat[i, j]
        annot.iat[i, j] = "" if pd.isna(val) else f"{val:+.1f}"

# 9) Tracé
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10), sharex=True, sharey=True)

# Heatmap des écarts (Prédit − Réel)
sns.heatmap(
    pivot_diff, ax=axes[0],
    cmap="RdBu_r", center=0, vmin=-vlim, vmax=vlim,
    annot=annot, fmt="", linewidths=.3, linecolor='white',
    cbar_kws={"label": "Écart (Prédit − Réel)"}
)
axes[0].set_title("Écart mensuel AQI — Prédit − Réel (toute la période)")
axes[0].set_xlabel("")
axes[0].set_ylabel("Année")

# Heatmap de l’erreur absolue (optionnel mais utile)
sns.heatmap(
    pivot_abs, ax=axes[1],
    cmap="YlOrRd", vmin=0, vmax=np.nanpercentile(pivot_abs.values, 95),
    annot=True, fmt=".1f", linewidths=.3, linecolor='white',
    cbar_kws={"label": "Erreur absolue"}
)
axes[1].set_title("Erreur absolue mensuelle de l’AQI (|Prédit − Réel|)")
axes[1].set_xlabel("Mois")
axes[1].set_ylabel("Année")

plt.tight_layout()
plt.show()

# (optionnel) Petit tableau récap pour export
ecarts_tab = (merged.assign(Mois=lambda d: d['Mois_num'].map({i+1:m for i,m in enumerate(mois_labels)}))
                    [['Année','Mois','AQI_Reel_Moyen','AQI_Predit_Moyen','Ecart','AbsErr']])
# display(ecarts_tab.round(2))
# ecarts_tab.round(2).to_csv("ecarts_aqi_mensuels.csv", index=False)
```

```python
# === Cycle diurne : aire colorée par classe (AQI prédit) ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Vérifs
if 'df_test' not in globals():
    raise NameError("df_test est introuvable. Exécute d'abord la cellule qui le crée.")
if 'classify_aqi' not in globals():
    raise NameError("La fonction classify_aqi(...) est introuvable. Exécute la cellule où elle est définie.")
for col in ['Datetime', 'AQI_Predicted']:
    if col not in df_test.columns:
        raise ValueError(f"Colonne manquante dans df_test : {col}")

# 1) Moyenne par heure (0..23)
tmp = df_test[['Datetime', 'AQI_Predicted']].dropna().copy()
tmp['Heure'] = pd.to_datetime(tmp['Datetime']).dt.hour

stats = (tmp.groupby('Heure')['AQI_Predicted']
           .mean()
           .reindex(range(24)))  # on s'assure d'avoir 24 heures

# Remplir si certaines heures manquent
stats = stats.interpolate(limit_direction='both')  # interpole entre heures manquantes
y = stats.rolling(3, center=True, min_periods=1).mean().to_numpy()  # lissage léger
hours = np.arange(24)

# 2) Palette par classe (mêmes codes que tes heatmaps)
palette_classes = {
    "Excellent": "#2ECC71",
    "Bon":       "#A3E4D7",
    "Moyen":     "#F1C40F",
    "Médiocre":  "#E67E22",
    "Dangereux": "#E74C3C",
}

# 3) Tracé : on colore l'aire segment par segment (heure → heure+1)
fig, ax = plt.subplots(figsize=(10, 5))

for h in range(23):
    xs = np.linspace(h, h+1, 20)                      # segment horaire
    ys = np.interp(xs, [h, h+1], [y[h], y[h+1]])      # interpolation linéaire
    cls = classify_aqi(float((y[h] + y[h+1]) / 2.0))  # classe du segment (moyenne des 2 points)
    color = palette_classes.get(cls, "#95A5A6")
    ax.fill_between(xs, ys, 0, color=color, alpha=0.35, linewidth=0)

# Segment final (23→24) à hauteur constante
xs = np.linspace(23, 24, 20)
ys = np.full_like(xs, y[23], dtype=float)
cls = classify_aqi(float(y[23]))
ax.fill_between(xs, ys, 0, color=palette_classes.get(cls, "#95A5A6"), alpha=0.35, linewidth=0)

# Courbe par dessus
ax.plot(hours, y, marker='o', linewidth=2, color='black', label='AQI moyen par heure')

# Guidages / axes
for thr in [25, 50, 75, 100]:
    ax.axhline(thr, ls='--', lw=0.8, color='gray', alpha=0.35)
ax.set_xticks(range(0, 24))
ax.set_xlim(0, 23.99)
ax.grid(True, alpha=0.25)

ax.set_title("Évolution moyenne de l’AQI prédit sur 24 h (aire colorée par classe)")
ax.set_xlabel("Heure de la journée")
ax.set_ylabel("AQI prédit (moyenne)")

# Légende des classes
legend_handles = [Patch(facecolor=c, label=cls) for cls, c in palette_classes.items()]
ax.legend(handles=legend_handles, title="Classe AQI", loc='upper right', frameon=False)

plt.tight_layout()
plt.show()
```

V.2 Application Streamlit

```python
# -*- coding: utf-8 -*-
"""
App Streamlit : Prédiction d'AQI en temps réel (version enrichie)
- Affiche AQI numérique + catégoriel (badge)
- Jauge (gauge matplotlib) + barre d’échelle colorée
- Historique de session (courbe)
- Mode modèle notebook (best_model.pkl + features.pkl) ou secours CO/NO₂
"""
import os, time
import numpy as np
import pandas as pd
import streamlit as st
import matplotlib.pyplot as plt
from matplotlib import colors

try:
    import joblib
except Exception:
    joblib = None

# -----------------------------
# 0) Seuils & utilitaires AQI
# -----------------------------
def classify_aqi(aqi_value: float) -> str:
    if aqi_value <= 25:   return "Excellent"
    elif aqi_value <= 50: return "Bon"
    elif aqi_value <= 75: return "Moyen"
    elif aqi_value <= 100:return "Médiocre"
    else:                 return "Dangereux"

CLASS_COLORS = {
    "Excellent": "#2ECC71",  # vert
    "Bon":       "#A3E4D7",  # vert clair
    "Moyen":     "#F1C40F",  # jaune
    "Médiocre":  "#E67E22",  # orange
    "Dangereux": "#E74C3C",  # rouge
}

# conversions approx (25°C, 1 atm)
def co_mg_m3_to_ppm(co_mg_m3):   return np.nan if pd.isna(co_mg_m3) else co_mg_m3 * 0.873
def no2_ug_m3_to_ppb(no2_ug_m3): return np.nan if pd.isna(no2_ug_m3) else no2_ug_m3 * 0.5229

# Règles EPA linéaires
CO_BREAKS = [(0.0,4.4,0,50),(4.5,9.4,51,100),(9.5,12.4,101,150),(12.5,15.4,151,200),(15.5,30.4,201,300),(30.5,40.4,301,400),(40.5,50.4,401,500)]
NO2_BREAKS= [(0,53,0,50),(54,100,51,100),(101,360,101,150),(361,649,151,200),(650,1249,201,300),(1250,1649,301,400),(1650,2049,401,500)]
def linear_aqi(Cp, Clow, Chigh, Ilow, Ihigh): return (Ihigh-Ilow)/(Chigh-Clow)*(Cp-Clow)+Ilow
def aqi_from_breaks(Cp, breaks):
    if pd.isna(Cp): return np.nan
    for Clow, Chigh, Ilow, Ihigh in breaks:
        if Clow <= Cp <= Chigh: return linear_aqi(Cp, Clow, Chigh, Ilow, Ihigh)
    return 500.0 if Cp > max(b[1] for b in breaks) else np.nan
def aqi_epa_from_co_no2(co_mg_m3, no2_ug_m3):
    co_ppm  = co_mg_m3_to_ppm(co_mg_m3)      # 8h approx
    no2_ppb = no2_ug_m3_to_ppb(no2_ug_m3)    # 1h
    sub_co  = aqi_from_breaks(co_ppm, CO_BREAKS)
    sub_n2  = aqi_from_breaks(no2_ppb, NO2_BREAKS)
    return float(np.nanmax([sub_co, sub_n2]))

# -----------------------------
# 1) Chargement du modèle
# -----------------------------
MODEL_PATH, FEATS_PATH = "best_model.pkl", "features.pkl"
model, features = None, None
model_status = "Aucun modèle trouvé — mode secours disponible"
if os.path.exists(MODEL_PATH) and os.path.exists(FEATS_PATH) and joblib is not None:
    try:
        model = joblib.load(MODEL_PATH)
        feat_raw = joblib.load(FEATS_PATH)
        if isinstance(feat_raw, (list, tuple, np.ndarray)):
            features = list(feat_raw)
        elif isinstance(feat_raw, dict) and "features" in feat_raw:
            features = list(feat_raw["features"])
        else:
            features = list(feat_raw)
        model_status = "Modèle chargé ✓"
    except Exception as e:
        model, features = None, None
        model_status = f"Échec chargement modèle : {e}"

# -----------------------------
# 2) VISUELS (indicateurs)
# -----------------------------
def colored_badge(label: str):
    color = CLASS_COLORS.get(label, "#BDC3C7")
    st.markdown(
        f"""<div style="display:inline-block;padding:.35rem .6rem;border-radius:999px;background:{color};
               color:#1B2631;font-weight:600;">{label}</div>""",
        unsafe_allow_html=True,
    )

def aqi_progress(aqi: float, max_ref: float = 150.0):
    pct = int(max(0.0, min(100.0, (aqi / max_ref) * 100.0)))
    st.progress(pct, text=f"AQI ~ {aqi:.1f} (≈{pct}%)")

def gauge_matplotlib(aqi: float, aqi_max: float = 150.0):
    """Jauge demi-cercle simple (0 -> aqi_max). Couleurs par zones AQI."""
    fig, ax = plt.subplots(figsize=(4, 2.2), subplot_kw={'projection': 'polar'})
    ax.set_theta_offset(np.pi)     # start at left
    ax.set_theta_direction(-1)     # clockwise
    ax.set_ylim(0, 10)
    ax.set_axis_off()

    # zones colorées (angles)
    zones = [
        (0, 25, "#2ECC71"),   # Excellent
        (25, 50, "#A3E4D7"),  # Bon
        (50, 75, "#F1C40F"),  # Moyen
        (75,100, "#E67E22"),  # Médiocre
        (100, aqi_max, "#E74C3C")  # Dangereux
    ]
    def val2theta(v): return np.pi * (1 - min(v, aqi_max) / aqi_max)
    for v0, v1, col in zones:
        t0, t1 = val2theta(v0), val2theta(v1)
        ax.bar(x=(t0+t1)/2, height=10, width=(t0 - t1), bottom=0, color=col, edgecolor="white", linewidth=1)

    # aiguille
    theta = val2theta(aqi)
    ax.plot([theta, theta], [0, 10], color="black", linewidth=2)
    ax.scatter([theta], [10], color="black", s=12)

    # ticks principaux
    for v in [0, 25, 50, 75, 100, aqi_max]:
        t = val2theta(v)
        ax.text(t, 11.3, f"{int(v)}", ha="center", va="center", fontsize=9)
    ax.text(np.pi/2, -2.5, "Jauge AQI", ha="center", va="center", fontsize=10, fontweight="bold")
    st.pyplot(fig, clear_figure=True)

def scale_bar(aqi: float):
    """Barre horizontale graduée vert->rouge + curseur à la valeur AQI."""
    fig, ax = plt.subplots(figsize=(6.5, 1.2))
    ax.axis("off")
    # dégradé
    cmap_grad = colors.LinearSegmentedColormap.from_list(
        "aqi_grad", ["#2ECC71", "#A3E4D7", "#F1C40F", "#E67E22", "#E74C3C"], N=256)
    grad = np.linspace(0, 1, 256).reshape(1, -1)
    ax.imshow(grad, aspect="auto", cmap=cmap_grad, extent=[0, 125, 0.2, 0.8])  # jusqu'à 125, au-delà = danger
    # curseur
    x = float(np.clip(aqi, 0, 125))
    ax.plot([x, x], [0.1, 0.9], color="black", linewidth=2)
    # ticks labels
    for x_lab, txt in zip([12.5, 37.5, 62.5, 87.5, 112.5], ["Excellent", "Bon", "Moyen", "Médiocre", "Dangereux"]):
        ax.text(x_lab, -0.05, txt, ha="center", va="top", fontsize=9)
    ax.set_title("Échelle AQI (curseur à la valeur prédite)", fontsize=10)
    st.pyplot(fig, clear_figure=True)

# -----------------------------
# 3) UI
# -----------------------------
st.set_page_config(page_title="AQI Temps Réel", page_icon="🌫️", layout="centered")
st.title("🌫️ Prédiction d’AQI en temps réel")
st.caption("Utilise le modèle entraîné s’il est présent ; sinon, calcul de secours via CO & NO₂.")
st.sidebar.header("Configuration")
st.sidebar.info(model_status)
mode = st.sidebar.radio("Mode", ["Modèle du notebook", "Calcul de secours (CO & NO₂)"],
                        index=0 if model is not None else 1)

st.markdown("---")

# -----------------------------
# 4) Prédiction + VISUELS
# -----------------------------
if mode == "Modèle du notebook" and model is not None and features is not None:
    st.subheader("🧠 Entrées du modèle")

    def default_for(feat: str) -> float:
        f = feat.lower()
        if any(k in f for k in ["temp"]): return 20.0
        if any(k in f for k in ["rh", "humid"]): return 50.0
        if any(k in f for k in ["co","no2","c6h6","nmhc","o3","so2","pm"]): return 1.0
        return 0.0

    input_vals, cols = {}, st.columns(2)
    for i, feat in enumerate(features):
        with cols[i % 2]:
            input_vals[feat] = st.number_input(feat, value=float(default_for(feat)), step=0.1, format="%.3f")

    X = pd.DataFrame([input_vals]).reindex(columns=features, fill_value=0.0)

    try:
        aqi_val = float(model.predict(X)[0])
        aqi_cat = classify_aqi(aqi_val)

        # KPI + badge
        k1, k2 = st.columns([1, 2])
        with k1: st.metric("AQI", f"{aqi_val:.1f}")
        with k2:
            st.markdown("**Classe :**")
            colored_badge(aqi_cat)

        # Visuels
        scale_bar(aqi_val)
        gauge_matplotlib(aqi_val)
        aqi_progress(aqi_val)

        with st.expander("Détails des entrées"):
            st.dataframe(X.T.rename(columns={0: "valeur"}))

        # Historique
        hist = st.session_state.get("history", [])
        hist.append({"timestamp": time.time(), "AQI": aqi_val, "classe": aqi_cat})
        st.session_state["history"] = hist[-200:]
        if st.checkbox("Afficher l'historique de cette session"):
            hdf = pd.DataFrame(st.session_state["history"])
            if not hdf.empty:
                hdf["timestamp"] = pd.to_datetime(hdf["timestamp"], unit="s")
                st.line_chart(hdf.set_index("timestamp")["AQI"])

    except Exception as e:
        st.error(f"Erreur de prédiction : {e}")
        st.info("Vérifie que best_model.pkl est un Pipeline (prétraitements inclus) et que features.pkl correspond.")

else:
    st.subheader("🧮 Calcul de secours (EPA-like)")
    c1, c2 = st.columns(2)
    with c1:
        co = st.number_input("CO(GT) [mg/m³]", value=1.0, min_value=0.0, step=0.1, format="%.2f")
    with c2:
        no2 = st.number_input("NO₂(GT) [µg/m³]", value=40.0, min_value=0.0, step=1.0, format="%.1f")

    aqi_val = aqi_epa_from_co_no2(co, no2)
    aqi_cat = classify_aqi(aqi_val)

    k1, k2 = st.columns([1, 2])
    with k1: st.metric("AQI", f"{aqi_val:.1f}")
    with k2:
        st.markdown("**Classe :**")
        colored_badge(aqi_cat)

    scale_bar(aqi_val)
    gauge_matplotlib(aqi_val)
    aqi_progress(aqi_val)

    with st.expander("Détails des conversions"):
        st.write(f"CO ≈ {co_mg_m3_to_ppm(co):.2f} ppm (8h, approx)")
        st.write(f"NO₂ ≈ {no2_ug_m3_to_ppb(no2):.1f} ppb (1h)")

st.markdown("---")
st.caption("💡 Astuce : depuis le notebook, exporte ton modèle : joblib.dump(best_model, 'best_model.pkl'); "
           "et la liste des features : joblib.dump(features, 'features.pkl').")
```

```python
# === Équation Lasso (approximation du best_model) ===
# Prérequis: scikit-learn==1.5.2, joblib, pandas, numpy

import joblib, numpy as np, pandas as pd
from pathlib import Path
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error

MODEL_PATH = Path("best_model.pkl")
X_PATH = Path("X.pkl")
Y_PATH = Path("y.pkl")

# 1) Charger objets
best_model = joblib.load(MODEL_PATH)
X = joblib.load(X_PATH)
y = joblib.load(Y_PATH)

# 2) Matrice et noms
if isinstance(X, pd.DataFrame):
    X_mat = X.values
    feature_names = list(X.columns)
else:
    X_mat = X
    feature_names = [f"f{i}" for i in range(X_mat.shape[1])]

# 3) Prédictions du modèle d'origine (lecture seule)
y_hat_rf = best_model.predict(X)

# 4) Standardisation + LassoCV pour approximer RF
scaler = StandardScaler(with_mean=True, with_std=True)
X_std = scaler.fit_transform(X_mat)
lasso = LassoCV(cv=5, random_state=42, n_alphas=100).fit(X_std, y_hat_rf)

# 5) Reprojection sur variables originales
sigmas, mus = scaler.scale_, scaler.mean_
b_coefs = lasso.coef_ / sigmas
b0 = lasso.intercept_ - np.sum(lasso.coef_ * mus / sigmas)

# 6) Équation lisible
terms = [f"AQI ≈ {b0:.8f}"]
for name, bi in zip(feature_names, b_coefs):
    if abs(bi) > 1e-10:
        s = "+" if bi >= 0 else "-"
        terms.append(f" {s} {abs(bi):.8f}·{name}")
equation = "".join(terms)

# 7) Fidélité de l'équation à votre modèle
y_hat_lasso = b0 + X_mat.dot(b_coefs)
r2_sur = r2_score(y_hat_rf, y_hat_lasso)
mae_sur = mean_absolute_error(y_hat_rf, y_hat_lasso)

# 8) (Optionnel) Écart vs vérité terrain y
r2_true = r2_score(y, y_hat_lasso)
mae_true = mean_absolute_error(y, y_hat_lasso)

# 9) Sorties
print("=== ÉQUATION LASSO (variables originales) ===")
print(equation)
print("\nFidélité à votre modèle : R² = {:.3f} | MAE = {:.3f}".format(r2_sur, mae_sur))
print("Vs vérité terrain (y)   : R² = {:.3f} | MAE = {:.3f}".format(r2_true, mae_true))

# Sauvegarder en fichiers texte (lecture/écriture uniquement, ne touche pas best_model.pkl)
Path("lasso_unstandardized_equation.txt").write_text(equation, encoding="utf-8")
Path("surrogate_metrics.txt").write_text(
    f"Fidelity to RF: R2={r2_sur:.4f}, MAE={mae_sur:.4f}\n"
    f"Vs true y: R2={r2_true:.4f}, MAE={mae_true:.4f}",
    encoding="utf-8"
)
print("\nÉcrit : lasso_unstandardized_equation.txt et surrogate_metrics.txt")
```
